```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
# Leemos los datos del archivo de planea y los almacenamos en la variable datos
datos = pd.read_csv('/content/sample_data/planea.csv', delimiter=',')
datos
datos.head()
```

```python
# Mostramos los datos de los registros del archivo
datos.head()
```

```python
# Mostramos la información general del archivo meramente para visualizar el numero de columnas, filas y tipo de datos.
datos.info()
```

```python
datos2= datos.drop(['nombre','turno','municipio'],axis=1)
datos2.head()
```

```python
# Mostramos lo que hay en la variable datos2.
datos2
```

```python
# NORMALIZACION DE DATOS
```

```python
# Con la ayuda de la funcion describe logramos visualizar que los datos no estan normalizados.
# Tenemos que normalizarlos para aplicar K-means esto quiere decir que esten en un valor entre 0-1.
datos2.describe()
```

```python
#normalizar datos
estandariza = StandardScaler()
datos_norm = estandariza.fit_transform(datos2)
datos_norm
```

```python
# APLICACION DE KMEANS
```

```python
# Hacemos el fit con los datos ya normalizados.
# ----NOTA----
# En este caso al momento de hacer el fit nos advierte que el valor de n_init cambia de 10 a ´auto´ por lo tanto, haremos los dos escenarios
# uno en donde no se modifica la variable n_init de manera explicita (Escenario1) y otro en donde si se modifica de manera explicita (Escenario2).
```

```python
# Escenario 1: Sin modificar la variable n_init explicitamente.
# Aplicamos el algoritmo K Means con un numero de agrupacion de 4 y la semilla 42.
kmeans = KMeans(n_clusters= 4, random_state= 42)
```

```python
kmeans.fit(datos_norm)
```

```python
datos['Grupo'] =kmeans.labels_ #los resultados del clustering
datos
```

```python
grupo1 = datos[(datos['Grupo'] == 1)]
grupo1.head()
```

```python
grupo2 = datos[(datos['Grupo'] == 2)]
grupo3 = datos[(datos['Grupo'] == 3)]
grupo4 = datos[(datos['Grupo'] == 4)]
```
