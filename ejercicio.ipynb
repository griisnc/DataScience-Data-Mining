```python
import pandas as ps
import numpy as np
import seaborn as sb

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import CategoricalNB

datos = ps.read_csv('/content/sample_data/titanic_NB.csv', delimiter=';')
datos
```

```python
from sklearn.model_selection import train_test_split

datos2 = datos[['pclass','survived','sex','edad','sibsp','parch']]
datos2
```

```python
datos2 = datos2.dropna()
```

```python
datos2
```

# Nueva secciÃ³n

```python
X_train, X_test = train_test_split(datos2,test_size=0.3,random_state=42)
```

```python
X_test
```

```python
y_train = X_train['survived']
y_test = X_test['survived']
X_train =X_train[['pclass','sex','edad','sibsp','parch']]
X_train
```

```python
X_test =X_test[['pclass','sex','edad','sibsp','parch']]
X_test
```

```python
ct = CategoricalNB()
  # para datos continuos hay que usar
  # gauss = GaussianNB()

```

```python
ct.fit(X_train,y_train) #funcion para hacer ajuste de bayes
```

```python
pronostico = ct.predict(X_test)
```

```python
pronostico
```

```python
print (ct.score(X_test,y_test))
```

```python
confusion_matrix(y_test,pronostico) #los de diagonal se suman para obtener la eficiencia del algoritmo (160+90) =250

#se suma todo (160+19+45+90) = 314 / 250 =0.79% = eficiencia
```
