```python
import pandas as ps
import numpy as np
import seaborn as sb


from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor

datos = ps.read_csv('/content/sample_data/diabetes_exa.csv', delimiter=',')
datos
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(datos.groupby('class').size()) #obteniendo la medida de la clasificacion de resultados de pacientes con diabetes
```

```python
X_train, X_test = train_test_split(datos, test_size=0.3, random_state=42)
X_train
```

```python
y_train = X_train['class']
y_test = X_test['class']
```

```python
X_train = X_train.drop(['class'],axis=1)
X_train
```

```python
X_test = X_test.drop(['class'],axis=1)
X_test
```

```python
ct = GaussianNB()
```

a) Obtén un clasificador en Python para la condición de Diabetes.

R= El clasificador es GaussianNB


```python
ct.fit(X_train,y_train)
```

```python
#se obtiene el pronostico de X_test
pronostico = ct.predict(X_test)
```

```python
pronostico
```

c) ¿Cuál es la eficiencia del modelo que obtuviste?
La eficiencia es del 73%


```python
#puntaje de exactitud del modelo de clasificación
print (ct.score(X_test,y_test))
```

d) ¿Cuántos verdaderos positivos y cuántos falsos negativos resultaron?
R= Se tuvieron 118 verdaderos negativos y 35 falsos negativos

```python
confusion_matrix(y_test,pronostico)
```
