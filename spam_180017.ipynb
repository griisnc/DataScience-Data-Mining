```python
import pandas as ps
import numpy as np
import seaborn as sb
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
with open('/content/sample_data/shortnames.txt') as lista: #abre el txt 'shortnames.txt'
    shortnames = lista.read().splitlines() #lee cada l√≠nea como un elemento de la lista
```

```python
shortnames
```

```python
#lee el txt 'spam.data2.txt'
spamData = ps.read_csv('/content/sample_data/spam.data2.txt', sep='\t', header=None)
```

```python
spamData.columns = shortnames #se agregan las columnas a spamData
spamData
```

```python
#spamData2 = spamData
spamData2 = spamData[[
'make', 'address', 'all', 'threeD', 'our', 'over', 'remove', 'internet', 'order', 'mail', 'receive', 'will', 'people', 'report', 'addresses', 'free', 'business', 'email', 'you', 'credit', 'your', 'font', 'n000', 'money', 'hp', 'hpl', 'george', 'n650', 'lab', 'labs', 'telnet', 'n857', 'data', 'n415', 'n85', 'technology', 'n1999', 'parts', 'pm', 'direct', 'cs', 'meeting', 'original', 'project', 're', 'edu', 'table', 'conference', 'semicolon', 'leftparen', 'leftbrack', 'bang', 'dollar', 'hash', 'crl.av', 'crl.long', 'crl.tot','yesno']]
```

```python
spamData2 = spamData2.dropna()
spamData2
```

```python
#se asignan datos de forma aleatoria a los conjuntos de prueba y entrenamiento
X_train, X_test = train_test_split(spamData2, test_size=0.3,random_state=42)
```

```python
#y_train y y_test toman la columna 'yesno'
y_train = X_train['yesno']
y_test = X_test['yesno']
```

```python
#se quita la columna 'yesno'
X_train = X_train[[
'make', 'address', 'all', 'threeD', 'our', 'over', 'remove', 'internet', 'order', 'mail', 'receive', 'will', 'people', 'report', 'addresses', 'free', 'business', 'email', 'you', 'credit', 'your', 'font', 'n000', 'money', 'hp', 'hpl', 'george', 'n650', 'lab', 'labs', 'telnet', 'n857', 'data', 'n415', 'n85', 'technology', 'n1999', 'parts', 'pm', 'direct', 'cs', 'meeting', 'original', 'project', 're', 'edu', 'table', 'conference', 'semicolon', 'leftparen', 'leftbrack', 'bang', 'dollar', 'hash', 'crl.av', 'crl.long', 'crl.tot']]
X_train
```

```python
#se quita la columna 'yesno'
X_test = X_test[[
'make', 'address', 'all', 'threeD', 'our', 'over', 'remove', 'internet', 'order', 'mail', 'receive', 'will', 'people', 'report', 'addresses', 'free', 'business', 'email', 'you', 'credit', 'your', 'font', 'n000', 'money', 'hp', 'hpl', 'george', 'n650', 'lab', 'labs', 'telnet', 'n857', 'data', 'n415', 'n85', 'technology', 'n1999', 'parts', 'pm', 'direct', 'cs', 'meeting', 'original', 'project', 're', 'edu', 'table', 'conference', 'semicolon', 'leftparen', 'leftbrack', 'bang', 'dollar', 'hash', 'crl.av', 'crl.long', 'crl.tot']]
X_test
```

```python
ct = GaussianNB()
```

```python
ct.fit(X_train,y_train)
```

```python
#se obtiene el pronostico de X_test
pronostico = ct.predict(X_test)
```

```python
pronostico
```

```python
#puntaje de exactitud del modelo de clasificaci√≥n
print (ct.score(X_test,y_test))
```
